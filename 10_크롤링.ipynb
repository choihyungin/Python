# get_html 함수

```python
def get_html(url):
    html = ''
    res = requests.get(url)
    if res.status_code == 200:
        res.encoding = None
        html = res.text
    return html
```

```python
import requests
result = get_html('http://ggoreb.com/python/html/data1.html')

s_idx = 0
e_idx = 0

while True:
    s_idx = result.find('<td>', e_idx)
    if s_idx == -1:
        break
    e_idx = result.find('</td>', s_idx+1)
    print(result[s_idx+4:e_idx])
    
```

```python
import time
#해외 여행 상세 목록 페이지 (실제 작업할 페이지 1 ~ 6)
time.sleep(2)
```

```python
# 해외여행 패키지 개수 확인
totalCnt = driver.find_element_by_id('totalAllCnt').text
print(totalCnt)
totalCnt = int(totalCnt)


# 전체 페이지 수를 확인 (그래야 반복문을 돌릴 수 있음)
totalPage = totalCnt // 10 + 1
if totalCnt % 10 == 0:
    totalPage -= 1
print(totalPage)
```

```python
for page in range(1, totalPage +1):
    driver.execute_script('searchModule.SetCategoryList(%d, '')' % page)
    
    
    # 1페이지로 이동
    time.slee(2)

    boxitems = driver.find_elements_by_css_selector('.oTravelBox .boxItem')
    for item in boxitems:
        print(item.find_element_by_class_name('proTit').text)
        print(item.find_element_by_css_selector('img').get_attribure('src'))
```
