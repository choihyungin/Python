```python
try:
    a = 2/0
    print(a)
    
except ZeroDivisionError as e:
    print(e, "예외 발생")
```

```python
# 예외가 발생하지 않았다면 try에 이어 else까지 실행하기

try:
    file = open('python.txt', 'r')
    
except FileNotFoundError as e:
    print(e, '예외발생')
    
else:
    file.read() 
    print(file) 
    file.close()
 
```

```python
# 예외 발생 여부와 상관 없이 무조건, 항상 실행

try:
    file = open('test.txt', 'r') 
    print('file open')
    text = file.read() print(text)
    text += 1
except:
    print('예외발생')
finally:
    file.close()
```

```python
try:
    file = open('text.txt', 'r')
except FileNotFoundError as e:
    print('예외발생', e)
else:
    text = file.read()
    print(text)
```
