```python
scoreList = [20, 55, 67, 82, 45, 33, 90, 87, 100, 25]
total = 0

for i in scoreList:
    if i >= 50:
        total += i
print(total)
```

```python
studentInfo = {'a':33, 'b':55, 'c':67, 'd':82, 'e':45,
             'f':33, 'g':90, 'h':87, 'i': 100, 'j':25}
total = 0

for student in studentInfo:
    if studentInfo[student] >= 80:
        total += studentInfo[student]
print(total)
```
